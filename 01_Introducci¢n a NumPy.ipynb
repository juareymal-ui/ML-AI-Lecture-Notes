## Arrays

Un **array** es una estructura de datos que consiste en una colección de elementos (valores o variables), cada uno identificado por al menos un índice o clave. Un array se almacena de modo que la posición de cada elemento se pueda calcular a partir de su tupla de índice mediante una fórmula matemática. El tipo más simple de array es un array lineal, también llamado array unidimensional.

En numpy:
* Cada dimensión se denomina **axis**
* El número de dimensiones se denomina **rank**
* La lista de dimensiones con su correspondiente longitud se denomina **shape**
* El número total de elementos (multiplicación de la longitud de las dimensiones) se denomina **size**

```python
import numpy as np
import matplotlib.pyplot as plt

def movimiento_proyectil():
    """
    Simula el movimiento parabólico de un proyectil
    """
    v0 = input("Introduce la velocidad inicial en m/s:" )  # velocidad inicial (m/s)
    angulo = input("Introduce el ángulo de lanzamiento en grados:" )   # ángulo de lanzamiento (grados)
    g = 9.81  # aceleración gravitacional (m/s²)
    
    # Convertir ángulo a radianes
    theta = np.radians(angulo)
    
    # Componentes de velocidad
    vx = v0 * np.cos(theta)
    vy = v0 * np.sin(theta)
    
    # Tiempo de vuelo
    t_vuelo = 2 * vy / g
    t = np.linspace(0, t_vuelo, 100)
    
    # Posición en función del tiempo
    x = vx * t
    y = vy * t - 0.5 * g * t**2
    
    print(f"Alcance máximo: {x[-1]:.2f} m")
    print(f"Altura máxima: {np.max(y):.2f} m")
    
    plt.figure(figsize=(10, 5))
    plt.plot(x, y)
    plt.xlabel('Distancia (m)')
    plt.ylabel('Altura (m)')
    plt.title('Trayectoria del Proyectil')
    plt.grid(True)
    plt.show()
movimiento_proyectil()
```

```python
def oscilador_armonico():
    """
    Simula un oscilador armónico simple
    """
    m = 1.0  # masa (kg)
    k = 10.0  # constante del resorte (N/m)
    omega = np.sqrt(k/m)  # frecuencia angular
    
    t = np.linspace(0, 10, 1000)
    x0 = 1.0  # posición inicial (m)
    v0 = 0.0  # velocidad inicial (m/s)
    
    # Solución analítica
    A = np.sqrt(x0**2 + (v0/omega)**2)
    phi = np.arctan2(-v0/omega, x0)
    x = A * np.cos(omega * t + phi)
    v = -A * omega * np.sin(omega * t + phi)
    
    # Energía
    E_k = 0.5 * m * v**2
    E_p = 0.5 * k * x**2
    E_total = E_k + E_p
    
    plt.figure(figsize=(12, 8))
    plt.subplot(2, 1, 1)
    plt.plot(t, x, label='Posición')
    plt.ylabel('Posición (m)')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(2, 1, 2)
    plt.plot(t, E_k, label='Energía Cinética')
    plt.plot(t, E_p, label='Energía Potencial')
    plt.plot(t, E_total, label='Energía Total', linestyle='--')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Energía (J)')
    plt.legend()
    plt.grid(True)
    plt.show()
oscilador_armonico()
```

```python

def campo_electrico_dipolo():
    """
    Calcula el campo eléctrico de un dipolo eléctrico
    """
    k = 8.99e9  # constante de Coulomb (N·m²/C²)
    q = 1e-9  # carga (C)
    d = 0.1  # separación entre cargas (m)
    
    # Crear malla
    x = np.linspace(-0.5, 0.5, 30)
    y = np.linspace(-0.5, 0.5, 30)
    X, Y = np.meshgrid(x, y)
    
    # Posiciones de las cargas
    pos_pos = np.array([d/2, 0])
    pos_neg = np.array([-d/2, 0])
    
    # Calcular campo eléctrico
    Ex = np.zeros_like(X)
    Ey = np.zeros_like(Y)
    
    for i in range(len(x)):
        for j in range(len(y)):
            r_pos = np.array([X[j, i], Y[j, i]]) - pos_pos
            r_neg = np.array([X[j, i], Y[j, i]]) - pos_neg
            
            r_pos_mag = np.linalg.norm(r_pos)
            r_neg_mag = np.linalg.norm(r_neg)
            
            if r_pos_mag > 0.01 and r_neg_mag > 0.01:
                E_pos = k * q * r_pos / r_pos_mag**3
                E_neg = k * (-q) * r_neg / r_neg_mag**3
                
                Ex[j, i] = E_pos[0] + E_neg[0]
                Ey[j, i] = E_pos[1] + E_neg[1]
    
    plt.figure(figsize=(10, 8))
    plt.streamplot(X, Y, Ex, Ey, density=1.5, color=np.sqrt(Ex**2 + Ey**2))
    plt.plot([pos_pos[0]], [pos_pos[1]], 'ro', markersize=10, label='+q')
    plt.plot([pos_neg[0]], [pos_neg[1]], 'bo', markersize=10, label='-q')
    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    plt.title('Campo Eléctrico de un Dipolo')
    plt.legend()
    plt.axis('equal')
    plt.colorbar(label='|E| (N/C)')
    plt.show()
campo_electrico_dipolo()
```

```python
def circuito_rlc():
    """
    Analiza un circuito RLC en serie
    """
    R = 100  # resistencia (Ω)
    L = 0.1  # inductancia (H)
    C = 1e-6  # capacitancia (F)
    V0 = 10  # voltaje inicial (V)
    
    # Frecuencia natural y factor de amortiguamiento
    omega0 = 1 / np.sqrt(L * C)
    gamma = R / (2 * L)
    omega_d = np.sqrt(omega0**2 - gamma**2)
    
    t = np.linspace(0, 0.01, 1000)
    
    # Corriente (caso subamortiguado)
    if gamma < omega0:
        I = (V0 / (omega_d * L)) * np.exp(-gamma * t) * np.sin(omega_d * t)
    
    # Voltajes
    V_R = R * I
    V_L = L * np.gradient(I, t)
    V_C = V0 - np.trapz(I, t, axis=0) / C
    
    plt.figure(figsize=(12, 6))
    plt.plot(t * 1000, I * 1000, label='Corriente (mA)')
    plt.plot(t * 1000, V_R, label='V_R (V)')
    plt.xlabel('Tiempo (ms)')
    plt.ylabel('Amplitud')
    plt.title('Circuito RLC - Respuesta Transitoria')
    plt.legend()
    plt.grid(True)
    plt.show()
circuito_rlc()
```

```python

def particula_en_caja():
    """
    Resuelve el problema de una partícula en una caja 1D
    """
    L = 1.0  # longitud de la caja (nm)
    n_max = 4  # número de estados a visualizar
    
    x = np.linspace(0, L, 1000)
    
    plt.figure(figsize=(12, 8))
    
    for n in range(1, n_max + 1):
        # Función de onda
        psi = np.sqrt(2/L) * np.sin(n * np.pi * x / L)
        
        # Densidad de probabilidad
        prob = psi**2
        
        plt.subplot(2, 1, 1)
        plt.plot(x, psi + n, label=f'n={n}')
        
        plt.subplot(2, 1, 2)
        plt.plot(x, prob + n, label=f'n={n}')
    
    plt.subplot(2, 1, 1)
    plt.xlabel('Posición (nm)')
    plt.ylabel('ψ(x) + n')
    plt.title('Funciones de Onda')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(2, 1, 2)
    plt.xlabel('Posición (nm)')
    plt.ylabel('|ψ(x)|² + n')
    plt.title('Densidad de Probabilidad')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
particula_en_caja()
```

```python
def oscilador_armonico_cuantico():
    """
    Calcula los niveles de energía del oscilador armónico cuántico
    """
    hbar = 1.054571817e-34  # constante de Planck reducida (J·s)
    m = 9.10938e-31  # masa del electrón (kg)
    omega = 1e15  # frecuencia angular (rad/s)
    
    # Niveles de energía
    n = np.arange(0, 10)
    E = hbar * omega * (n + 0.5)
    
    # Convertir a eV
    E_eV = E / 1.602176634e-19
    
    print("Niveles de energía del oscilador armónico cuántico:")
    for i, energia in enumerate(E_eV):
        print(f"n={i}: E = {energia:.6f} eV")
    
    plt.figure(figsize=(10, 6))
    plt.hlines(E_eV, 0, 1, colors='blue', linewidth=2)
    plt.ylabel('Energía (eV)')
    plt.title('Niveles de Energía - Oscilador Armónico Cuántico')
    plt.xlim(-0.1, 1.1)
    plt.xticks([])
    plt.grid(True, axis='y')
    plt.show()
oscilador_armonico_cuantico()
```

```python
def paquete_onda_gaussiano():
    """
    Simula la evolución temporal de un paquete de onda gaussiano
    """
    # Parámetros
    x = np.linspace(-50, 50, 1000)
    k0 = 2  # número de onda central
    sigma = 5  # ancho del paquete
    
    # Paquete de onda inicial
    psi = np.exp(-(x**2)/(4*sigma**2)) * np.exp(1j*k0*x)
    psi = psi / np.sqrt(np.trapz(np.abs(psi)**2, x))
    
    plt.figure(figsize=(12, 6))
    plt.plot(x, np.abs(psi)**2, label='|ψ(x)|²')
    plt.plot(x, np.real(psi), label='Re(ψ)', alpha=0.5)
    plt.plot(x, np.imag(psi), label='Im(ψ)', alpha=0.5)
    plt.xlabel('Posición')
    plt.ylabel('Amplitud')
    plt.title('Paquete de Onda Gaussiano')
    plt.legend()
    plt.grid(True)
    plt.show()
paquete_onda_gaussiano()
```

```python

def distribucion_maxwell_boltzmann():
    """
    Calcula la distribución de velocidades de Maxwell-Boltzmann
    """
    k_B = 1.380649e-23  # constante de Boltzmann (J/K)
    m = 4.65e-26  # masa de molécula de N2 (kg)
    T_values = [100, 300, 500, 1000]  # temperaturas (K)
    
    v = np.linspace(0, 2000, 1000)
    
    plt.figure(figsize=(10, 6))
    
    for T in T_values:
        f_v = 4 * np.pi * (m / (2 * np.pi * k_B * T))**(3/2) * v**2 * np.exp(-m * v**2 / (2 * k_B * T))
        plt.plot(v, f_v, label=f'T = {T} K')
        
        # Velocidad más probable
        v_prob = np.sqrt(2 * k_B * T / m)
        print(f"T = {T} K: v_más_probable = {v_prob:.2f} m/s")
    
    plt.xlabel('Velocidad (m/s)')
    plt.ylabel('f(v)')
    plt.title('Distribución de Maxwell-Boltzmann para N₂')
    plt.legend()
    plt.grid(True)
    plt.show()
distribucion_maxwell_boltzmann()
```

```python
def gas_ideal_presion():
    """
    Simula N partículas de un gas ideal y calcula la presión
    """
    N = 1000  # número de partículas
    L = 10  # tamaño del contenedor (m)
    m = 1e-26  # masa (kg)
    v_rms = 500  # velocidad rms (m/s)
    
    # Velocidades iniciales (distribución gaussiana)
    vx = np.random.normal(0, v_rms/np.sqrt(3), N)
    vy = np.random.normal(0, v_rms/np.sqrt(3), N)
    vz = np.random.normal(0, v_rms/np.sqrt(3), N)
    
    # Calcular presión
    v_squared_mean = np.mean(vx**2 + vy**2 + vz**2)
    V = L**3  # volumen
    P = N * m * v_squared_mean / (3 * V)
    
    print(f"Presión calculada: {P:.2e} Pa")
    print(f"Velocidad rms: {np.sqrt(v_squared_mean):.2f} m/s")
    
    # Histograma de velocidades
    v_total = np.sqrt(vx**2 + vy**2 + vz**2)
    
    plt.figure(figsize=(10, 6))
    plt.hist(v_total, bins=50, density=True, alpha=0.7, edgecolor='black')
    plt.xlabel('Velocidad (m/s)')
    plt.ylabel('Densidad de probabilidad')
    plt.title('Distribución de Velocidades de las Partículas')
    plt.grid(True)
    plt.show()
gas_ideal_presion()
```
